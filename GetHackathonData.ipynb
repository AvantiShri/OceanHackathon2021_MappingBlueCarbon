```python
import pandas as pd
```

```python
compiled = pd.read_csv('bluecarbondata_compiled.csv')
compiled.head()
```

```python
print(len(compiled))
print(len(compiled.dropna()))
print(len(compiled.dropna(subset=['Latitude','Longitude'])))
print(len(set(compiled['Latitude'])))
print(len(set(compiled['Longitude'])))
```

```python
compiled['latlon'] = [str(x['Latitude'])+'_'+str(x['Longitude']) for ix,x in compiled.iterrows()]
compiled.head()
```

```python
len(set(compiled['latlon']))
```

```python
unique = compiled.drop_duplicates(subset='latlon')
print(len(unique))
unique.head()
```

```python
from collections import Counter
Counter(unique['Ecosystem'])
```

```python
unique.to_csv('bluecarbondata_unique.csv', index=False)
unique[['Ecosystem','Latitude','Longitude']].to_csv('bluecarbondata_coordinates.csv',index=False)
```

## create train valid test dfs separately

```python
unique = pd.read_csv('bluecarbondata_unique.csv')
unique.head()
```

```python
train = unique[unique['dataset']=='train']
valid = unique[unique['dataset']=='valid']
test = unique[unique['dataset']=='test']
```

```python
from collections import Counter
print(Counter(unique['Ecosystem']))
print(Counter(train['Ecosystem']))
print(Counter(valid['Ecosystem']))
print(Counter(test['Ecosystem']))
```

```python
train[['Ecosystem','Latitude','Longitude']].to_csv('train_coordinates.csv',index=False)
valid[['Ecosystem','Latitude','Longitude']].to_csv('valid_coordinates.csv',index=False)
test[['Ecosystem','Latitude','Longitude']].to_csv('test_coordinates.csv',index=False)
```
